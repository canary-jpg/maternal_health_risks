## Exploring the Outliers
In statistics, an outlier is an observation point that is very different from other observations. An outlier can be due to data collection or recording mistakes, or due to the naturally high variability of data points. How to treat an outlier highly depends on our data or the type of analysis we want to perform. Outliers can greatly affect our models and can also be a valuable source of information, providing us with insights about specific behaviors.

There are many ways to discover outliers in our data. We can perform uni-variate analysis (using one variable analysis) or multi-variate analysis (using two or more variables). One of the simplest ways to detect an outlier is to inspect the data visually, by making box plots.

```python
sns.set_palette("Dark2")
sns.boxplot(maternal_risks);
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
sns.boxplot(y="BS", data=maternal_risks)
plt.subplot(1, 3, 2)
sns.boxplot(y="Age", data=maternal_risks)
plt.subplot(1, 3, 3)
sns.boxplot(y="BodyTemp", data=maternal_risks)

plt.tight_layout()
plt.show()
```

As we can see from the plots above, we have some points that are plotted well outside the box plot area and that greatly deviates from the rest of the population. Whether or not we want to remove or keep these observations greatly depends on the understanding of our data and the type of analysis to be performed. In this case, the points outside of our box plots might be the actual true data points and do not need to be removed.

## Skew Variables Transformation
Skewness refers to data distortion or symmetry/asymmetry of data distribution. When your data is skewed, the value of the mean and median is different. Skewness violates the normality assumption of some ML models, like for example, linear regression.

Now, I will create a pairplot to see a histogram of each column, in order to check if any of the variables are skewed.

```python
sns.pairplot(maternal_risks);
```

Visually, it looks like the BS (blood sugar) and the BodyTemp (body temperature) are highly skewed. They both have longer tails to the right, which is known as a positive skew. Let's calculate the skewness levels of both columns with the `skew()` method.

```python
#defining a limit above which we will log transform
skew_limit = 0.75

mask = maternal_risks.dtypes != object
num_cols = maternal_risks.columns[mask]

skew_values = maternal_risks[num_cols].skew()
```

```python
skew_cols = (skew_values.sort_values(ascending=False).to_frame().rename(columns={0:'Skew'}).query('abs(Skew) > {}'.format(skew_limit)))
skew_cols
```

The range of skewness for a fairly symmetrical bell curve distribution is between -0.5 and 0.5; moderate skewness is -0.5 and -1.0 and 0.5 and 1.0; and highly skewed distribution is < -1.0 and > 1.0. In our case, we have a skewness of blood sugar is ~ 2.26 and body temperature's skewness is ~1.75, which is considered highly skewed data. The skew for age is around 0.92, which is considered moderate skewness.
The distribution of the BodyTemp column makes it impossible to transform to a normal distribution (normal human body temperature is around 98 degrees and most rows will have the same value), so we will not change it. However we can transform the blood glucose and age columns, so they will look more normally distributed. There are three common transformations to make features be more normally distributed:
* Log 
* Square root
* Box cox

```python
maternal_transformed = maternal_risks.copy()

maternal_transformed["Age"] = maternal_risks["Age"].apply(np.log)
print(maternal_transformed["Age"].skew())
sns.displot(maternal_transformed.Age);
```

#### Box Cox
The Box cox transformation is a parameterized transformation that tries to get distributions "as close to a normal distribution as possible". 

```python
bc_result = boxcox(maternal_risks.BS)
boxcox_bs = pd.DataFrame(bc_result[0], columns=["BS"])
lambd = bc_result[1]
print(lambd)
```

```python
print(boxcox_bs.skew())
sns.displot(boxcox_bs);
```

```python
maternal_transformed["BS"] = boxcox_bs["BS"]
```

```python
maternal_transformed[["BS", "Age"]].skew().to_frame().rename(columns={0: "Skew"}).sort_values("Skew", ascending=False)
```

After performing the log and box cox transformations, the skewness of:
* blood glucose lowered from 2.26 to 0.26
* and age 0.92 to 0.2
now all our distributions are fairly normal.
